# Pymaceuticals
### Analysis

Capomulin was the only treatment which resulted in reducing the volume of tumors. Infubinol and Ketapril do not result in the reduction of tumor size and the volume of the tumors are indistinguishable from the placebo treatment group. Capomulin reduced the volume of tumors by almost 20% compared to the start of the treatment, while tumors increased in size by 50% in the placebo and comparably in the other treatment groups.

Both Capomulin and Infubinol reduced the instance of metastatic sites over the course of the 45 day treatment, whereas Ketapril did not result in a reduction as compared to the placebo. The placebo treatment group developed on average 3.27 metastatic sites, and the Ketapril treatment group developed an average of 3.36 metastatic sites. The Infubinol treatment group averaged 2.11 metastatic sites, and the Capomulin treatment resulted in the greatest reduction over the placebo with 1.48 metastatic sites. 

Capomulin was the only treatment that improved survival rates over the placebo treatment. Both placebo and Ketapril resulted in a 45% survival rate, while the Infubinol treatment group had a 35% survival rate. Survival rate in the Capomulin treatment group was 85%.

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.axes as ax
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
drug_data_file = "mouse_drug_data.csv"
trial_data_file = "clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
drug_data=pd.read_csv(drug_data_file)
trial_data=pd.read_csv(trial_data_file)

# Combine the data into a single dataset
merge_data=pd.merge(trial_data, drug_data, on="Mouse ID", how="left")

# Display the data table for preview
merge_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame
groupby_drug=merge_data.groupby(["Drug", "Timepoint"])
tumor_size_by_drug_timepoint=groupby_drug["Tumor Volume (mm3)"]
mean_tumor_size_by_drug_timepoint=pd.DataFrame({"Mean Tumor Volume (mm3)":tumor_size_by_drug_timepoint.mean()}).reset_index()
mean_tumor_size_by_drug_timepoint
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
# Preview DataFrame
sem_tumor_size_by_drug_timepoint=pd.DataFrame({"SEM Tumor Volume (mm3)":tumor_size_by_drug_timepoint.sem()}).reset_index()
sem_tumor_size_by_drug_timepoint.head()

```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked
mean_tumor_size_timepoint_by_drug=mean_tumor_size_by_drug_timepoint.pivot(index="Timepoint",columns="Drug", values="Mean Tumor Volume (mm3)")
sem_tumor_size_timepoint_by_drug=sem_tumor_size_by_drug_timepoint.pivot(index="Timepoint",columns="Drug", values="SEM Tumor Volume (mm3)")

mean_tumor_size_timepoint_by_drug.head()
```

```python
# Generate the Plot (with Error Bars)
x_axis=list(mean_tumor_size_timepoint_by_drug.index)
capomulin_tumor_resp=plt.errorbar(x_axis, mean_tumor_size_timepoint_by_drug["Capomulin"], yerr=sem_tumor_size_timepoint_by_drug["Capomulin"], linewidth=0.5, marker='o',color='red',label="Capomulin")
infubinol_tumor_resp=plt.errorbar(x_axis, mean_tumor_size_timepoint_by_drug["Infubinol"], yerr=sem_tumor_size_timepoint_by_drug["Infubinol"], linewidth=0.5, marker='^',color='blue',label="Infubinol")
ketapril_tumor_resp=plt.errorbar(x_axis, mean_tumor_size_timepoint_by_drug["Ketapril"], yerr=sem_tumor_size_timepoint_by_drug["Ketapril"], linewidth=0.5, marker='s', color='green', label="Ketapril")
placebo_tumor_resp=plt.errorbar(x_axis, mean_tumor_size_timepoint_by_drug["Placebo"], yerr=sem_tumor_size_timepoint_by_drug["Placebo"], linewidth=0.5, marker='d', color='black', label="Placebo")

plt.ylim(33, 73)
plt.grid()
plt.legend(loc="upper left")
plt.title("Tumor Response by Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")

# Save the Figure
plt.savefig("Images/tumor_response_by_treatment.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame
met_site_by_drug_timepoint=groupby_drug["Metastatic Sites"]
mean_met_site_by_drug_timepoint=pd.DataFrame({"Mean Metastatic Sites":met_site_by_drug_timepoint.mean()}).reset_index()
pd.DataFrame(met_site_by_drug_timepoint.mean()).head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame
sem_met_site_by_drug_timepoint=pd.DataFrame({"SEM Metastatic Sites":met_site_by_drug_timepoint.sem()}).reset_index()
pd.DataFrame(met_site_by_drug_timepoint.sem()).head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked
mean_met_site_timepoint_by_drug=mean_met_site_by_drug_timepoint.pivot(index="Timepoint",columns="Drug", values="Mean Metastatic Sites")
sem_met_site_timepoint_by_drug=sem_met_site_by_drug_timepoint.pivot(index="Timepoint",columns="Drug", values="SEM Metastatic Sites")

mean_met_site_timepoint_by_drug.head()
```

```python
# Generate the Plot (with Error Bars)
# Save the Figure
# Show the Figure

capomulin_met_sites=plt.errorbar(x_axis, mean_met_site_timepoint_by_drug["Capomulin"], yerr=sem_met_site_timepoint_by_drug["Capomulin"], linewidth=0.5, marker='o',color='red',label="Capomulin")
infubinol_met_sites=plt.errorbar(x_axis, mean_met_site_timepoint_by_drug["Infubinol"], yerr=sem_met_site_timepoint_by_drug["Infubinol"], linewidth=0.5, marker='^',color='blue',label="Infubinol")
ketapril_met_sites=plt.errorbar(x_axis, mean_met_site_timepoint_by_drug["Ketapril"], yerr=sem_met_site_timepoint_by_drug["Ketapril"], linewidth=0.5, marker='s', color='green', label="Ketapril")
placebo_met_sites=plt.errorbar(x_axis, mean_met_site_timepoint_by_drug["Placebo"], yerr=sem_met_site_timepoint_by_drug["Placebo"], linewidth=0.5, marker='d', color='black', label="Placebo")

plt.grid()
plt.legend(loc="upper left")
plt.title("Metastatic Sites by Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Metastatic Sites")

# Save the Figure
plt.savefig("Images/metastatic_sites_by_treatment.png")
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# Convert to DataFrame
# Preview DataFrame

survival_by_drug_timepoint=groupby_drug["Mouse ID"]
survival_by_drug_timepoint=pd.DataFrame({"Mouse Count":survival_by_drug_timepoint.count()}).reset_index()
survival_by_drug_timepoint.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview the Data Frame
survival_timepoint_by_drug=survival_by_drug_timepoint.pivot(index="Timepoint",columns="Drug", values="Mouse Count")
survival_timepoint_by_drug.head()
```

```python
# Generate the Plot (Accounting for percentages)

capomulin_survival=plt.errorbar(x_axis, survival_timepoint_by_drug["Capomulin"]/25*100, linewidth=0.5, marker='o',color='red',label="Capomulin")
infubinol_survival=plt.errorbar(x_axis, survival_timepoint_by_drug["Infubinol"]/25*100, linewidth=0.5, marker='^',color='blue',label="Infubinol")
ketapril_survival=plt.errorbar(x_axis, survival_timepoint_by_drug["Ketapril"]/25*100, linewidth=0.5, marker='s', color='green', label="Ketapril")
placebo_survival=plt.errorbar(x_axis, survival_timepoint_by_drug["Placebo"]/25*100, linewidth=0.5, marker='d', color='black', label="Placebo")

plt.grid()
plt.legend(loc="lower left")
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")

# Save the Figure
plt.savefig("Images/survival_by_treatment.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
percent_change=mean_tumor_size_timepoint_by_drug.loc[45,:]/mean_tumor_size_timepoint_by_drug.loc[0,:]*100-100

# Display the data to confirm
percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple
tumor_pct_changes=(percent_change["Capomulin"],percent_change["Infubinol"],percent_change["Ketapril"],percent_change["Placebo"])

# Splice the data between passing and failing drugs
x_axis=np.arange(4)
x_labels=["Capomulin", "Infubinol", "Ketapril", "Placebo"]
tick_locations=[value for value in x_axis]
plt.bar(x_axis, tumor_pct_changes, color=['green' if tumor_pct_changes[x]<0 else 'red' for x in x_axis])


# Orient widths. Add labels, tick marks, etc. 
plt.axhline(y=0, color='black')
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
plt.xticks(tick_locations, x_labels)
plt.ylim(-30, 70)


# Use functions to label the percentages of changes
def autolabel(values_list):
    for idx, value in enumerate(values_list):
        if value > 0:
            plt.text(idx-0.25, 5, f'{round(value,2)}%', color='white', fontsize=12, fontweight='bold')
        else:
            plt.text(idx-0.35, -10, f'{round(value,2)}%', color='white', fontsize=12, fontweight='bold')


# Call functions to implement the function calls
autolabel(tumor_pct_changes)

# Save the Figure
plt.savefig("Images/pct_change_by_treatment.png")

# Show the Figure
plt.show()
```
