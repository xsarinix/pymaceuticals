```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
drug_data_file = "mouse_drug_data.csv"
trial_data_file = "clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
drug_data=pd.read_csv(drug_data_file)
trial_data=pd.read_csv(trial_data_file)

# Combine the data into a single dataset
merge_data=pd.merge(trial_data, drug_data, on="Mouse ID", how="left")

# Display the data table for preview
merge_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame
groupby_drug=merge_data.groupby(["Drug", "Timepoint"])
tumor_size_by_drug_timepoint=groupby_drug["Tumor Volume (mm3)"]
mean_tumor_size_by_drug_timepoint=pd.DataFrame({"Mean Tumor Volume (mm3)":tumor_size_by_drug_timepoint.mean()}).reset_index()
mean_tumor_size_by_drug_timepoint
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
# Preview DataFrame
sem_tumor_size_by_drug_timepoint=pd.DataFrame({"SEM Tumor Volume (mm3)":tumor_size_by_drug_timepoint.sem()}).reset_index()
sem_tumor_size_by_drug_timepoint

```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked
mean_tumor_size_timepoint_by_drug=mean_tumor_size_by_drug_timepoint.pivot(index="Timepoint",columns="Drug", values="Mean Tumor Volume (mm3)")
mean_tumor_size_timepoint_by_drug
sem_tumor_size_timepoint_by_drug=sem_tumor_size_by_drug_timepoint.pivot(index="Timepoint",columns="Drug", values="SEM Tumor Volume (mm3)")
print(len(sem_tumor_size_timepoint_by_drug["Capomulin"]))
print(len(mean_tumor_size_timepoint_by_drug["Capomulin"]))
print(len(list(mean_tumor_size_timepoint_by_drug["Capomulin"].index.values)))
x_axis=np.array(mean_tumor_size_timepoint_by_drug.index.values.tolist())
mean_tumor_size_timepoint_by_drug["Capomulin"].tolist()
```

```python
# Generate the Plot (with Error Bars)

capomulin_handle=plt.errorbar(x_axis, mean_tumor_size_timepoint_by_drug["Capomulin"], yerr=sem_tumor_size_timepoint_by_drug["Capomulin"], linewidth=0.5, marker='o',color='red',label="Capomulin")
infubinol_handle=plt.errorbar(x_axis, mean_tumor_size_timepoint_by_drug["Infubinol"], yerr=sem_tumor_size_timepoint_by_drug["Infubinol"], linewidth=0.5, marker='^',color='blue',label="Infubinol")
ketapril_handle=plt.errorbar(x_axis, mean_tumor_size_timepoint_by_drug["Ketapril"], yerr=sem_tumor_size_timepoint_by_drug["Ketapril"], linewidth=0.5, marker='s', color='green', label="Ketapril")
placebo_handle=plt.errorbar(x_axis, mean_tumor_size_timepoint_by_drug["Placebo"], yerr=sem_tumor_size_timepoint_by_drug["Placebo"], linewidth=0.5, marker='d', color='black', label="Placebo")

plt.ylim(33, 73)
plt.grid()
plt.legend(loc="upper left")
plt.title("Tumor Response by Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")

# Save the Figure
plt.savefig("Images/tumor_response_by_treatment.png")
```

```python
# Show the Figure
plt.show()
```
